```
import numpy as np
import matplotlib.pyplot as plt
samples = 100
C1 = [-50,-10]
R1 = 10
X1 = (np.random.rand(samples,2)*2*R1-R1)+C1
Y1 = np.zeros(samples)
C2 = [0,30]
R2 = 7
X2 = (np.random.rand(samples,2)*2*R2-R2)+C2
Y2 = np.zeros(samples) + 1
C3 = [50,-50]
R3 = 10
X3 = (np.random.rand(samples,2)*2*R3-R3)+C3
Y3 = np.zeros(samples) + 2
X = np.append(X3,np.append(X2,X1,axis=0),axis=0)
y = np.append(Y3,np.append(Y2,Y1,axis=0),axis=0)
plt.plot(X[:, 0][y == 0], X[:, 1][y == 0], 'g^')
plt.plot(X[:, 0][y == 1], X[:, 1][y == 1], 'b*')
plt.plot(X[:, 0][y == 2], X[:, 1][y == 2], 'r+')
plt.show()
```

```
y1 = (y==0)*2-1
y2 = (y==1)*2-1
y3 = (y==2)*2-1
plt.plot(X[:, 0][y1 == 1], X[:, 1][y1 == 1], 'g^')
plt.plot(X[:, 0][y2 == 1], X[:, 1][y2 == 1], 'b*')
plt.plot(X[:, 0][y3 == 1], X[:, 1][y3 == 1], 'r+')
plt.show()
```

```
def train(X,yk):
    n_samples = X.shape[0]
    n_features = X.shape[1]
    weights = np.zeros((n_features + 1,))
    Xs = np.concatenate([X, np.ones((n_samples, 1))], axis=1)
    for i in range(100):
        for j in range(n_samples):
            if yk[j] * np.dot(weights, Xs[j, :]) <= 0:
                weights += yk[j] * Xs[j, :]
    return weights
weights1 = train(X,y1)
weights2 = train(X,y2)
weights3 = train(X,y3)
```

```
plt.plot(X[:, 0][y1 == 1], X[:, 1][y1 == 1], 'g^')
plt.plot(X[:, 0][y2 == 1], X[:, 1][y2 == 1], 'b*')
plt.plot(X[:, 0][y3 == 1], X[:, 1][y3 == 1], 'r+')
x_values = [np.min(X[:,1])*2, np.max(X[:,1])*2]
y_values1 = [(-weights1[2]-weights1[0]*x_values[0])/weights1[1], (-weights1[2]-weights1[0]*x_values[1])/weights1[1]]
y_values2 = [(-weights2[2]-weights2[0]*x_values[0])/weights2[1], (-weights2[2]-weights2[0]*x_values[1])/weights2[1]]
y_values3 = [(-weights3[2]-weights3[0]*x_values[0])/weights3[1], (-weights3[2]-weights3[0]*x_values[1])/weights3[1]]
plt.plot(x_values, y_values1)
plt.plot(x_values, y_values2)
plt.plot(x_values, y_values3)
plt.show()
```
